```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import plotly.express as px
import scipy.stats as stats
import plotly.offline as pyo
import plotly.offline as pyo
```

```python
df_postalcode_5 = pd.read_pickle("postalcode5/postcode5_1_normal.pkl")

df_postalcode_5.reset_index(drop=True, inplace=True)
df_postalcode_5.head(8)
```

```python
def label_damage (row):
   if row['target'] == 1:
      return 'Rainwater damage'
   if row['target'] == 0:
      return 'No rainwater damage'

df_postalcode_5.past3hours = df_postalcode_5.past3hours / 10
df_postalcode_5['label'] = df_postalcode_5.apply(lambda row: label_damage(row), axis=1)

df_postalcode_5.head(2)
```

# Map Visualization

```python
color_scale = [(0, 'blue'), (1,'red')]

fig = px.scatter_mapbox(df_postalcode_5, 
                        lat="lat", 
                        lon="lng", 
                        hover_name="group", 
                        hover_data=["target","date","past3hours", 'zipcode', 'group'],
                        color="target",
                        color_continuous_scale=color_scale,
                        zoom=8, 
                        height=800,
                        width=800)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
